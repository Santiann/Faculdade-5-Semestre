```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread('cafe.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
img2 = img.copy()

h = img2.shape[0]
w = img2.shape[1]
threshold = 90

for i in range(0, h):
    for j in range(0, w):
        if(img2[i,j] > threshold):
            img2[i,j] = 0
        else:
            img2[i,j] = 255

plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
#metodo = cv2.THRESH_BINARY
metodo = cv2.THRESH_BINARY_INV
threshold = 90

ret,img3 = cv2.threshold(img, threshold, 255, metodo)

plt.imshow(img3, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
#metodo = cv2.THRESH_BINARY
metodo = cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU

ret,img3 = cv2.threshold(img, 0, 255, metodo)

plt.imshow(img3, cmap='gray', vmin=0, vmax=255)
plt.show()

print(ret)
```

```python
elementoEstruturante = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
#elementoEstruturante = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
#elementoEstruturante = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5))
elementoEstruturante
```

```python
metodo = cv2.THRESH_BINARY_INV
ret,imgb = cv2.threshold(img, 90, 255, metodo)

img4 = cv2.erode(imgb, elementoEstruturante, iterations = 1)

plt.imshow(imgb, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.imshow(img4, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
metodo = cv2.THRESH_BINARY_INV
ret,imgb = cv2.threshold(img, 90, 255, metodo)

img4 = cv2.dilate(imgb, elementoEstruturante, iterations = 2)

plt.imshow(imgb, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.imshow(img4, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
metodo = cv2.THRESH_BINARY_INV
ret,imgb = cv2.threshold(img, 90, 255, metodo)

img4 = cv2.morphologyEx(imgb, cv2.MORPH_OPEN, elementoEstruturante)

plt.imshow(imgb, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.imshow(img4, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
metodo = cv2.THRESH_BINARY_INV
ret,imgb = cv2.threshold(img, 90, 255, metodo)

img4 = cv2.morphologyEx(imgb, cv2.MORPH_CLOSE, elementoEstruturante)

plt.imshow(imgb, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.imshow(img4, cmap='gray', vmin=0, vmax=255)
plt.show()
```
